```python
import time,random
import re
def get_orgId(stock_code,zwjc=""):
    # if stock_code[:3] in SH_LIST:
    #     return 'gssh0'+str(stock_code)
    # elif stock_code[:3] in SZ_LIST:
    #     return "gssz0"+str(stock_code)

    def remain_hanzi(str_):
        return re.sub('[a-zA-Z Ａ]','',str_)


    r = requests.post(
                    'http://www.cninfo.com.cn/new/information/topSearch/query',
                    {'keyWord': stock_code},
                    )
    try:
        # all_ = r.content.decode('utf-8')
        all_ = r.json()
        print(all_)
        for content in all_:
            if remain_hanzi(content['zwjc']) == remain_hanzi(zwjc):
                res = content['orgId']
                break
            elif ('退市' in content['zwjc']) and (content['category']=='A股'):
                res = content['orgId']
                break
        else:
            print("Not found!")
            print(f"Current_code:【 {stock_code} 】")
            print(f"Current_name:【 {zwjc} 】")
            select = input("请输入选中的字典，如0为第一个")
            res = all_[int(select)]['orgId']

    except Exception as e :
        print(e)
        print(f"Current_code:【 {stock_code} 】")
        time.sleep(1)
        return None
    return res
```

```python
SZ_LIST = ['300','301','000','001','002','003']
SH_LIST = ['601','603','605','688','689','600']

code_df = pd.read_csv('./code_orgId.csv',dtype=str,)
code = code_df.set_index('code')

def get_query_dict(stock_code):
    stock_code = str(stock_code)
    orgId = code.loc[stock_code]['orgId']
    if stock_code[:3] in SZ_LIST:
        # sz创业
            res = {
                "stock":str(stock_code)+","+orgId,
                "column":"szse",
                'plate': 'sz',# ''

            }
    elif stock_code[:3] in SH_LIST:
        # sh创业
            res = {
                "stock":str(stock_code)+","+orgId,
                "column":"sse",
                'plate': 'sh',# ''

            }
    return res
```
